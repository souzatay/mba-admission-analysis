###  Análise Univariada das Variáveis Categóricas

```python
import matplotlib.pyplot as plt

columns = ["gender", "international", "major", "race"]
titles = ["Distribuição de Gênero", "Distribuição de Nacionalidade", "Distribuição de Majors", "Distribuição de Raça"]

fig, axes = plt.subplots(2, 2, figsize=(8, 8))

# Percorrendo as colunas e criando os gráficos
for i, ax in enumerate(axes.flatten()):
    counts = df[columns[i]].value_counts()
    percentages = counts / counts.sum() * 100

    ax.pie(percentages, labels=counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
    ax.set_title(titles[i])

# Ajustando layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Substituir NaN por "No Admit"
df["admission"] = df["admission"].fillna("No Admit")
counts = df["admission"].value_counts()

# Obtendo as contagens
admit = counts.get("Admit", 0)
no_admit = counts.get("No Admit", 0)
wait_list = counts.get("Waitlist", 0)

count = admit + no_admit + wait_list
print(f"Admitidos: {admit}, Não Admitidos: {no_admit}, Lista de espera: {wait_list}")

# Criando lista de categorias e valores
categories = ["Admitidos", "Não Admitidos", "Lista de Espera"]
values = [(admit/count*100), (no_admit/count*100), (wait_list/count)*100]

# Criando o gráfico de barras

plt.figure(figsize=(8, 5))  # Ajustando o tamanho do gráfico
plt.grid(linestyle='-', linewidth=0.5)
bars = plt.bar(categories, values, color=sns.color_palette("pastel"))

# Adicionando os valores acima das barras
for bar in bars:
    height = bar.get_height()  # Obtém a altura da barra (valor da categoria)
    plt.text(bar.get_x() + bar.get_width()/2, height, f"{height:.2f}",  # Formato com 2 casas decimais
             ha='center', va='bottom', fontsize=12, fontweight='bold')

# Adicionando legenda manualmente
plt.legend(bars, categories, title="Status da Candidatura")
# Adicionando rótulos e título
plt.ylabel("Porcentagem")
# Exibir o gráfico
plt.show()
```

```python

# Contando o total de candidatos locais (True) e não locais (False)
total_non_local = df["international"].sum()  # Candidatos locais (True)
total_local = (~df["international"]).sum()  # Candidatos não locais (False)

# Contando quantos locais e não locais foram admitidos
admitted_non_local = ((df["international"] == True) & (df["admission"] == "Admit")).sum()
admitted_local = ((df["international"] == False) & (df["admission"] == "Admit")).sum()

# Calculando as porcentagens
percent_local = (admitted_local / total_local) * 100 if total_local > 0 else 0
percent_non_local = (admitted_non_local / total_non_local) * 100 if total_non_local > 0 else 0

print(f"Porcentagem de candidatos locais admitidos: {percent_local:.2f}%")
print(f"Porcentagem de candidatos não locais admitidos: {percent_non_local:.2f}%")

# Criando o gráfico de barras
categories = ["Locais", "Não Locais"]
values = [percent_non_local, percent_local]
colors = ["blue", "purple"]  # Definindo cores para cada categoria

plt.figure(figsize=(8, 5))  # Ajustando o tamanho do gráfico
bars = plt.bar(categories, values, color= sns.color_palette("pastel"))

# Adicionando os valores acima das barras com 2 casas decimais
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f"{height:.2f}%",  
             ha='center', va='bottom', fontsize=12, fontweight='bold')

# Adicionando rótulos e título
plt.ylabel("Porcentagem de Admitidos (%)")
plt.title("Porcentagem de Candidatos Admitidos por Origem")

# Exibir o gráfico
plt.show()
```

```python
df_non_local = df[df["international"] == True]
gpa_non_local_sum = df_non_local["gpa"].sum()

df_local = df[df["international"] == False]
gpa_local_sum = df_local["gpa"].sum()


categories = ["Locais", "Não Locais"]
values = [gpa_non_local_sum/total_non_local, gpa_local_sum/total_local]
colors = ["blue", "purple"]  # Definindo cores para cada categoria

plt.figure(figsize=(8, 5))  # Ajustando o tamanho do gráfico
bars = plt.bar(categories, values, color= sns.color_palette("pastel"))

# Adicionando os valores acima das barras com 2 casas decimais
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f"{height:.2f}",  
             ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title("Comparação GPA por Origem")
plt.ylabel("Média GPA")
```

```python
gmat_non_local_sum = df_non_local["gmat"].sum()
gmat_local_sum = df_local["gmat"].sum()

categories = ["Locais", "Não Locais"]
values = [gmat_non_local_sum/total_non_local, gmat_local_sum/total_local]
colors = ["blue", "purple"]  # Definindo cores para cada categoria

plt.figure(figsize=(8, 5))  # Ajustando o tamanho do gráfico
bars = plt.bar(categories, values, color= sns.color_palette("pastel"))

# Adicionando os valores acima das barras com 2 casas decimais
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f"{height:.2f}",  
             ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title("Comparação GMAT por Origem")
plt.ylabel("Média GMAT")
```

### Teste de hipotese para comparar as medias GPA e GMAT entre participantes locais e não locais

Uma das operações estatísticas mais elementares consiste em comparar dois grupos de dados. Nesse contexto, o teste T de Student se destaca como uma ferramenta poderosa e popular

# Verificando relação entre GPA e GMAT para candidatos aprovados

```python
df_admitidos = df[df["admission"] == "Admit"]

sns.lmplot(data=df_admitidos, x="gpa", y="gmat", 
           scatter_kws={"alpha":0.5}, line_kws={"color":"red"}, ci=None)

plt.title("Relação entre GPA e GMAT para candidatos admitidos")
plt.show()
```

Observa-se no gráfico uma tendência positiva: candidatos com GPA mais alto parecem ter, em média, GMAT mais alto.

```python
from scipy.stats import spearmanr

corr_spearman, p_value_spearman = spearmanr(df_admitidos["gpa"], df_admitidos["gmat"])
print(f"Correlação de Spearman: {corr_spearman:.4f}, Valor-p: {p_value_spearman:.4f}")
```

Interpretação: 

Correlação: 0.5184 → Isso indica uma correlação moderada positiva entre GPA e GMAT. Ou seja, candidatos admitidos com um GPA mais alto tendem a ter uma pontuação maior no GMAT.

Valor-p: 0.0000 → Como o p-valor é menor que 0.05, a correlação encontrada é estatisticamente significativa. Isso significa que é muito improvável que essa relação tenha ocorrido por acaso

## Proporção de candidatos por industria 

```python

df["admission_binary"] = df["admission"].apply(lambda x: 1 if x == "Admit" else 0)

# Calcular a taxa de admissão por indústria
industry_admission = df.groupby("major")["admission_binary"].mean().sort_values(ascending=False)
# Criar um gráfico de barras

plt.figure(figsize=(12,6))
sns.barplot(x=industry_admission.index, y=industry_admission.values, palette=  sns.color_palette("pastel"))
plt.xticks(rotation=45)
plt.xlabel("Indústria")
plt.ylabel("Proporção de Admitidos")
plt.title("Proporção de Candidatos Admitidos por Indústria")
plt.show()
```

Observa-se que a proporção de candidatos nas diferentes áreas é aproximadamente a mesma, indicando que não existe nenhuma indústria em que a proporção seja significativamente maior.
