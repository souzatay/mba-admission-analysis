###  AnÃ¡lise Univariada das VariÃ¡veis CategÃ³ricas

```python
import matplotlib.pyplot as plt

columns = ["gender", "international", "major", "race"]
titles = ["DistribuiÃ§Ã£o de GÃªnero", "DistribuiÃ§Ã£o de Nacionalidade", "DistribuiÃ§Ã£o de Majors", "DistribuiÃ§Ã£o de RaÃ§a"]

fig, axes = plt.subplots(2, 2, figsize=(8, 8))

# Percorrendo as colunas e criando os grÃ¡ficos
for i, ax in enumerate(axes.flatten()):
    counts = df[columns[i]].value_counts()
    percentages = counts / counts.sum() * 100

    ax.pie(percentages, labels=counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
    ax.set_title(titles[i])

# Ajustando layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Substituir NaN por "No Admit"
df["admission"] = df["admission"].fillna("No Admit")
counts = df["admission"].value_counts()

# Obtendo as contagens
admit = counts.get("Admit", 0)
no_admit = counts.get("No Admit", 0)
wait_list = counts.get("Waitlist", 0)

count = admit + no_admit + wait_list
print(f"Admitidos: {admit}, NÃ£o Admitidos: {no_admit}, Lista de espera: {wait_list}")

# Criando lista de categorias e valores
categories = ["Admitidos", "NÃ£o Admitidos", "Lista de Espera"]
values = [(admit/count*100), (no_admit/count*100), (wait_list/count)*100]

# Criando o grÃ¡fico de barras

plt.figure(figsize=(8, 5))  # Ajustando o tamanho do grÃ¡fico
plt.grid(linestyle='-', linewidth=0.5)
bars = plt.bar(categories, values, color=sns.color_palette("pastel"))

# Adicionando os valores acima das barras
for bar in bars:
    height = bar.get_height()  # ObtÃ©m a altura da barra (valor da categoria)
    plt.text(bar.get_x() + bar.get_width()/2, height, f"{height:.2f}",  # Formato com 2 casas decimais
             ha='center', va='bottom', fontsize=12, fontweight='bold')

# Adicionando legenda manualmente
plt.legend(bars, categories, title="Status da Candidatura")
# Adicionando rÃ³tulos e tÃ­tulo
plt.ylabel("Porcentagem")
# Exibir o grÃ¡fico
plt.show()
```

```python

# Contando o total de candidatos locais (True) e nÃ£o locais (False)
total_non_local = df["international"].sum()  # Candidatos locais (True)
total_local = (~df["international"]).sum()  # Candidatos nÃ£o locais (False)

# Contando quantos locais e nÃ£o locais foram admitidos
admitted_non_local = ((df["international"] == True) & (df["admission"] == "Admit")).sum()
admitted_local = ((df["international"] == False) & (df["admission"] == "Admit")).sum()

# Calculando as porcentagens
percent_local = (admitted_local / total_local) * 100 if total_local > 0 else 0
percent_non_local = (admitted_non_local / total_non_local) * 100 if total_non_local > 0 else 0

print(f"Porcentagem de candidatos locais admitidos: {percent_local:.2f}%")
print(f"Porcentagem de candidatos nÃ£o locais admitidos: {percent_non_local:.2f}%")

# Criando o grÃ¡fico de barras
categories = ["Locais", "NÃ£o Locais"]
values = [percent_non_local, percent_local]
colors = ["blue", "purple"]  # Definindo cores para cada categoria

plt.figure(figsize=(8, 5))  # Ajustando o tamanho do grÃ¡fico
bars = plt.bar(categories, values, color= sns.color_palette("pastel"))

# Adicionando os valores acima das barras com 2 casas decimais
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f"{height:.2f}%",  
             ha='center', va='bottom', fontsize=12, fontweight='bold')

# Adicionando rÃ³tulos e tÃ­tulo
plt.ylabel("Porcentagem de Admitidos (%)")
plt.title("Porcentagem de Candidatos Admitidos por Origem")

# Exibir o grÃ¡fico
plt.show()
```

```python
df_non_local = df[df["international"] == True]
gpa_non_local_sum = df_non_local["gpa"].sum()

df_local = df[df["international"] == False]
gpa_local_sum = df_local["gpa"].sum()


categories = ["Locais", "NÃ£o Locais"]
values = [gpa_non_local_sum/total_non_local, gpa_local_sum/total_local]
colors = ["blue", "purple"]  # Definindo cores para cada categoria

plt.figure(figsize=(8, 5))  # Ajustando o tamanho do grÃ¡fico
bars = plt.bar(categories, values, color= sns.color_palette("pastel"))

# Adicionando os valores acima das barras com 2 casas decimais
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f"{height:.2f}",  
             ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title("ComparaÃ§Ã£o GPA por Origem")
plt.ylabel("MÃ©dia GPA")
```

```python
gmat_non_local_sum = df_non_local["gmat"].sum()
gmat_local_sum = df_local["gmat"].sum()

categories = ["Locais", "NÃ£o Locais"]
values = [gmat_non_local_sum/total_non_local, gmat_local_sum/total_local]
colors = ["blue", "purple"]  # Definindo cores para cada categoria

plt.figure(figsize=(8, 5))  # Ajustando o tamanho do grÃ¡fico
bars = plt.bar(categories, values, color= sns.color_palette("pastel"))

# Adicionando os valores acima das barras com 2 casas decimais
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f"{height:.2f}",  
             ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title("ComparaÃ§Ã£o GMAT por Origem")
plt.ylabel("MÃ©dia GMAT")
```

### Teste de hipotese para comparar as medias GPA e GMAT entre participantes locais e nÃ£o locais

Uma das operaÃ§Ãµes estatÃ­sticas mais elementares consiste em comparar dois grupos de dados. Nesse contexto, o teste T de Student se destaca como uma ferramenta poderosa e popular

# Verificando relaÃ§Ã£o entre GPA e GMAT para candidatos aprovados

```python
df_admitidos = df[df["admission"] == "Admit"]

sns.lmplot(data=df_admitidos, x="gpa", y="gmat", 
           scatter_kws={"alpha":0.5}, line_kws={"color":"red"}, ci=None)

plt.title("RelaÃ§Ã£o entre GPA e GMAT para candidatos admitidos")
plt.show()
```

Observa-se no grÃ¡fico uma tendÃªncia positiva: candidatos com GPA mais alto parecem ter, em mÃ©dia, GMAT mais alto.

```python
from scipy.stats import spearmanr

corr_spearman, p_value_spearman = spearmanr(df_admitidos["gpa"], df_admitidos["gmat"])
print(f"CorrelaÃ§Ã£o de Spearman: {corr_spearman:.4f}, Valor-p: {p_value_spearman:.4f}")
```

InterpretaÃ§Ã£o: 

CorrelaÃ§Ã£o: 0.5184 â†’ Isso indica uma correlaÃ§Ã£o moderada positiva entre GPA e GMAT. Ou seja, candidatos admitidos com um GPA mais alto tendem a ter uma pontuaÃ§Ã£o maior no GMAT.

Valor-p: 0.0000 â†’ Como o p-valor Ã© menor que 0.05, a correlaÃ§Ã£o encontrada Ã© estatisticamente significativa. Isso significa que Ã© muito improvÃ¡vel que essa relaÃ§Ã£o tenha ocorrido por acaso

## ProporÃ§Ã£o de candidatos por industria 

```python

df["admission_binary"] = df["admission"].apply(lambda x: 1 if x == "Admit" else 0)

# Calcular a taxa de admissÃ£o por indÃºstria
industry_admission = df.groupby("major")["admission_binary"].mean().sort_values(ascending=False)
# Criar um grÃ¡fico de barras

plt.figure(figsize=(12,6))
sns.barplot(x=industry_admission.index, y=industry_admission.values, palette=  sns.color_palette("pastel"))
plt.xticks(rotation=45)
plt.xlabel("IndÃºstria")
plt.ylabel("ProporÃ§Ã£o de Admitidos")
plt.title("ProporÃ§Ã£o de Candidatos Admitidos por IndÃºstria")
plt.show()
```

Observa-se que a proporÃ§Ã£o de candidatos nas diferentes Ã¡reas Ã© aproximadamente a mesma, indicando que nÃ£o existe nenhuma indÃºstria em que a proporÃ§Ã£o seja significativamente maior.
